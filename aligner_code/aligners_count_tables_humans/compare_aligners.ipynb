```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
bwa = pd.read_csv('bwa_human_count_table.csv')
bwa = bwa.rename(columns={'HLIVER_01_bspc_A03_1': 'bwa'})
bwa

```

```python
kallisto = pd.read_csv('kallisto_human_count_table.csv')
kallisto = kallisto.rename(columns={'HLIVER_01_bspc_A03_1': 'kallisto'})
kallisto
```

```python
star = pd.read_csv('star_human_count_table.csv')
star = star.rename(columns={'HLIVER_01_bspc_A03_1': 'star'})
star
```

```python
# Star vs kallisto
plt.scatter(star.star, kallisto.kallisto)
plt.xlim([0, 10000])
plt.ylim([0,10000])
plt.show()
```

```python
# Star vs bwa
plt.scatter(star.star, bwa.bwa)
plt.xlim([0, 10000])
plt.ylim([0,10000])
plt.show()
```

```python
# Bwa vs kallisto
plt.scatter(bwa.bwa, kallisto.kallisto)
plt.xlim([0, 10000])
plt.ylim([0,10000])
plt.show()
```

```python
#merge tables
conc = bwa.merge(kallisto, on='Geneid')
conc = conc.merge(star, on='Geneid')
```
