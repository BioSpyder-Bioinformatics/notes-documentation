```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
d = pd.read_csv("EPA0001_summary_stats.csv", header=0)
print(d.head())
```

```python
fig , ax = plt.subplots(1,1,figsize=(7,5))
sns.scatterplot(data=d, x='#UMI', y='Average detected genes', hue='Condition', style='Index', size='Mapping rate')
plt.xlabel('#UMI',fontweight='bold')
plt.ylabel('Average Detected genes',fontweight='bold')
plt.legend(loc='lower right')
sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
```

```python
d_cond1 = d[d['Mapping rate']>40]

fig , ax = plt.subplots(1,1,figsize=(7,5))
sns.scatterplot(data=d_cond1, x='#UMI', y='Average detected genes', label="0uL Exo")

plt.xlabel('#UMI',fontweight='bold')
plt.ylabel('Average Detected genes',fontweight='bold')
plt.legend(loc='lower right')
```

```python
d_av = d.groupby('Condition').mean()
#d_av.to_csv("BIOS2568_500_average_stats.csv")

d_std = d.groupby('Condition').std()
#d_std.to_csv("BIOS2568_500_std_stats.csv")

d_sum = d.groupby('Condition').sum()
#d_sum.to_csv("BIOS2568_500_sum_stats.csv")

d_sem = d.groupby('Condition').sem()
#d_sem.to_csv("BIOS2568_700_sem_stats.csv")

final_table = pd.concat([d_av,d_sum,d_std,d_sem],axis=0)
final_table.to_csv("BIOS2710_2000_summary.csv")
```

```python

# load S1 data
s1 = pd.read_csv("S01_S1/human_umi_count.csv", header=0)
s1.set_index(s1.columns[0], inplace=True)
x_s1 = []
y_s1 = []
for column in s1.columns:
    x_s1.append(s1[column].sum())
    y_s1.append(s1[column][s1[column]>0].shape[0])
    
# load S2 data
s2 = pd.read_csv("S02_S2/human_umi_count.csv", header=0)
s2.set_index(s2.columns[0], inplace=True)
x_s2 = []
y_s2 = []
for column in s2.columns:
    x_s2.append(s2[column].sum())
    y_s2.append(s2[column][s2[column]>0].shape[0])
    
# load s3 data
s3 = pd.read_csv("S03_S3/human_umi_count.csv", header=0)
s3.set_index(s3.columns[0], inplace=True)
x_s3 = []
y_s3 = []
for column in s3.columns:
    x_s3.append(s3[column].sum())
    y_s3.append(s3[column][s3[column]>0].shape[0])
```

```python
fig , ax = plt.subplots(1,1,figsize=(8,5))
#sns.scatterplot(data=d, x='#UMI', y='Average detected genes', hue='Condition',s=15)
#plt.scatter(x_bios2256,y_bios2256, color='red',label='BIOS2256', s=15, alpha=0.5)
plt.scatter(x_s1,y_s1, color='blue',label='S1', s=15, alpha=0.5)
plt.scatter(x_s2,y_s2, color='green',label='S2', s=15,alpha=0.5)
plt.scatter(x_s3,y_s3, color='purple',label='S3', s=15,alpha=0.5)
plt.xlabel("# mapped UMIs", fontweight='bold')
plt.ylabel("# Detected genes", fontweight='bold')

plt.legend(loc='lower right')
fig.savefig("umi_gene_comparison.jpg", dpi=300)
```
